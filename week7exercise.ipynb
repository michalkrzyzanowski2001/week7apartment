```python
#Q2:
import numpy as np
import numpy_financial as npf
import matplotlib.pyplot as plt
pv = 120000 * 1.05**5
freq = 12
rate = 0.12
years = 5

rate /= freq
nper = years * freq
periods = np.arange(1,nper+1,dtype=int)
interest_equal = pv + 100*np.around(npf.ipmt(rate,periods+1,nper,pv),2)
print(f"Monthly deposit value: {interest_equal[0]} PLN")
```

```python
#Q3:
appartment_value = np.linspace(120000,pv,60)
plt.plot(interest_equal,label='Deposit value')
plt.plot(appartment_value,label='Appartment value')
plt.legend()
plt.xlabel('# of periods')
plt.ylabel('PLN')
```
